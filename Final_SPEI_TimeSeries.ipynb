```python
import ee
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

districts = ee.FeatureCollection(
    'projects/ee-gc-gis/assets/DistrictShapefiles'
)

Kongwa = districts.filter(
    ee.Filter.eq('District_N', 'Kongwa')
)

center = Kongwa.geometry().centroid(10)
coords = center.coordinates().getInfo()

lon0 = coords[0]
lat0 = coords[1]

print(f"Kongwa center: lon {lon0:.3f}, lat {lat0:.3f}")

rings = {
    'Kongwa (~0.25°)': 0.25,
    'Ring 1 (~0.5°)': 0.5,
    'Ring 2 (~1.0°)': 1.0,
    'Ring 3 (~2.0°)': 2.0,
}


def box(half_width):
    return ee.Geometry.Rectangle([
        lon0 - half_width,
        lat0 - half_width,
        lon0 + half_width,
        lat0 + half_width
    ])


speiCol = (
    ee.ImageCollection('CSIC/SPEI/2_10')
    .filterDate('2002-01-01', '2023-01-01')
)


def extract(geom):

    def per_img(img):
        stats = (
            img.select([
                'SPEI_03_month',
                'SPEI_12_month',
                'SPEI_48_month'
            ])
            .reduceRegion(
                reducer=ee.Reducer.mean(),
                geometry=geom,
                scale=55000,
                bestEffort=True,
                maxPixels=1e9
            )
        )

        return ee.Feature(None, {
            'date': img.date().format('YYYY-MM-dd'),
            'SPEI_3': stats.get('SPEI_03_month'),
            'SPEI_12': stats.get('SPEI_12_month'),
            'SPEI_48': stats.get('SPEI_48_month')
        })

    fc = ee.FeatureCollection(speiCol.map(per_img))

    rows = fc.getInfo()['features']
    records = [feature['properties'] for feature in rows]

    d = pd.DataFrame(records)

    d['date'] = pd.to_datetime(d['date'])

    for column in ['SPEI_3', 'SPEI_12', 'SPEI_48']:
        d[column] = pd.to_numeric(
            d[column],
            errors='coerce'
        )

    return (
        d.sort_values('date')
        .reset_index(drop=True)
    )


series = {}

for name, half_width in rings.items():
    print(f"Extracting {name}...")
    series[name] = extract(box(half_width))

print("Done extracting all rings.")


plots = [
    ('SPEI_3', 'SPEI-3'),
    ('SPEI_12', 'SPEI-12'),
    ('SPEI_48', 'SPEI-48')
]


for band, title in plots:

    fig, ax = plt.subplots(figsize=(15, 7))

    for name in rings:
        d = series[name]

        ax.plot(
            d['date'],
            d[band],
            linewidth=1.3,
            label=name
        )

    ax.axhline(
        0,
        color='gray',
        linestyle='--',
        linewidth=0.8
    )

    ax.axvline(
        pd.Timestamp('2015-01-01'),
        color='purple',
        linestyle=':',
        linewidth=2,
        label='Preliminary study 2015'
    )

    ax.axvspan(
        pd.Timestamp('2018-01-01'),
        pd.Timestamp('2020-12-31'),
        color='green',
        alpha=0.12,
        label='MMT trial period (2018–2020)'
    )

    ax.axvline(
        pd.Timestamp('2018-01-01'),
        color='green',
        linestyle=':',
        linewidth=1.5
    )

    ax.axvline(
        pd.Timestamp('2020-12-31'),
        color='green',
        linestyle=':',
        linewidth=1.5
    )

    ax.set_title(
        f'{title}: Kongwa and Surrounding Regions (2002–2022)',
        fontsize=14,
        fontweight='bold'
    )

    ax.set_xlabel('Year')
    ax.set_ylabel(f'{title} (standardized)')

    ax.xaxis.set_major_locator(
        mdates.YearLocator(2)
    )

    ax.xaxis.set_major_formatter(
        mdates.DateFormatter('%Y')
    )

    all_values = pd.concat(
        [
            series[name][band]
            for name in rings
        ],
        ignore_index=True
    ).dropna()

    if not all_values.empty:
        ymin = np.floor(all_values.min() * 4) / 4
        ymax = np.ceil(all_values.max() * 4) / 4

        ax.set_yticks(
            np.arange(
                ymin - 0.25,
                ymax + 0.50,
                0.25
            )
        )

        ax.set_ylim(
            ymin - 0.25,
            ymax + 0.25
        )

    ax.grid(
        True,
        axis='both',
        alpha=0.3
    )

    ax.legend(
        loc='upper center',
        bbox_to_anchor=(0.5, -0.12),
        ncol=3,
        fontsize=9,
        frameon=True
    )

    plt.tight_layout()

    plt.savefig(
        f'{title.replace("-", "_")}_Kongwa_surrounding_regions_2002_2022.png',
        dpi=300,
        bbox_inches='tight'
    )

    plt.show()
```
